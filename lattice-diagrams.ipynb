```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = np.meshgrid([i for i in range(-4, 5)], [i for i in range(-4, 5)])
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'o', markersize=4, color=(0.5, 0.1, 0.4))
ax.arrow(0, 0, 0, 1, color=(0.4, 0.6, 0.1), width=0.01, head_width=0.12, length_includes_head=True)
ax.arrow(0, 0, 1, 0, color=(0.4, 0.6, 0.1), width=0.01, head_width=0.12, length_includes_head=True)
ax.arrow(0, 0, 2, 3, color=(0.1, 0.4, 0.6), width=0.001, head_width=0.12, length_includes_head=True)
ax.arrow(0, 0, 3, 4, color=(0.1, 0.4, 0.6), width=0.001, head_width=0.12, length_includes_head=True)
ax.arrow(0, 0, 0, 4, color='0', width=0.000005, head_width=0.1, length_includes_head=True)
ax.arrow(0, 0, 4, 0, color='0', width=0.000005, head_width=0.1, length_includes_head=True)
ax.arrow(0, 0, -4, 0, color='0', width=0.000005, head_width=0.1, length_includes_head=True)
ax.arrow(0, 0, 0, -4, color='0', width=0.000005, head_width=0.1, length_includes_head=True)

ax.set(title='Lattice Example')
ax.grid()

plt.savefig('lattice-examples')
plt.show()
```
